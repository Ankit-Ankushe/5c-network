```python
# create a data frame from the CSV file

import pandas as pd

df = pd.read_csv('weather_data.csv')
df
```

```python
# Get first 10 data from dataframe

df.head(10) # way 1

df = df[:10] # way 2
df
```

**Filtering the data**

```python
# Getting the data of whole row after appling filter
df[df['Temperature'] == df['Temperature'].max()]
```

```python
# Getting the data of only selected columns after applying filter
df[df['Temperature'] == df['Temperature'].min()][['EST','Temperature']]
```

```python
# Applying multiple filters
df[(df['Temperature']>20) & (df['Humidity']>50)][['EST','Temperature','Humidity']]
```

```python
# Set index to EST

df1 = df.set_index('EST')
df1
```

```python
# Reset Index

df.reset_index(inplace=True)
df
```

```python
# Dropping Entries whose values are NaN

df1 = df.dropna()
df1
```

```python
# Filling 0 in place of NaN

df1 = df.fillna(0)
df1
```

```python
# Filling NaN with some legit values

df1 = df.fillna({
    'Temperature': df['Temperature'].mean(),
    'WindSpeedMPH': df['WindSpeedMPH'].mean(),
    'Humidity': df['Humidity'].mean()
})
df1
```

```python
# Fill forward values in place of NaN

df1 = df.fillna(method='ffill')
df1
```

```python
# Fill Backward values in place of NaN

df1 = df.fillna(method='bfill')
df1
```

**Interpolate**

 The interpolate() method is used to fill missing values (NaN) by interpolating (estimating) values between known data points. Interpolation essentially estimates the missing data based on the surrounding data points.

```python
# Intrapolate the

df1 = df.interpolate()
df1
```

**Merge two data frames**

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner") # look at the common only : df1 intersection df2
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer") # df1 union df2
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left") # merge wrt to left arg : df1
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right") # merge wrt to right arg : df2
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True) # indicator of merge
df3
```

**If two dfs are having same columns but the value missmatch**

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```
